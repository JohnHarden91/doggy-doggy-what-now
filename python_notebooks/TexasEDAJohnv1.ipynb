```python

# Ensure the 'DateTime' column is in the correct format
Austindf['DateTime'] = pd.to_datetime(Austindf['DateTime'])

# Define the threshold date
covid_start_date = pd.to_datetime('2020-03-01')

# Add a new column initialized with 0
Austindf['COVID'] = 0

# Update the 'COVID' column using a loop
covid_counter = 0
for index, row in Austindf.iterrows():
    if row['DateTime'] >= covid_start_date:
        covid_counter += 1
    Austindf.at[index, 'COVID'] = covid_counter

print(Austindf["COVID"])
```

```python
plt.scatter(Austindf["TotalIntakes"], Austindf["COVID"])
plt.xlabel('Total Inttakes')
plt.ylabel('Days Since COVID start')
plt.title('Austin, Texas (2014 - 2024)')
```
